```python
import pandas as pd
train_full = pd.read_csv('https://raw.githubusercontent.com/NHANGIOI/Machine_Learning/refs/heads/main/house-prices-advanced-regression-techniques/train.csv')
x_test = pd.read_csv('https://raw.githubusercontent.com/NHANGIOI/Machine_Learning/refs/heads/main/house-prices-advanced-regression-techniques/test.csv')
x_train = train_full.copy().drop(columns = 'SalePrice')
y_train = train_full['SalePrice']
x_train.columns
```

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

numerical_cols = [col for col in x_train.columns if x_train[col].dtypes in ['int','float']]
caterical_cols = [col for col in x_train.columns if x_train[col].dtypes == 'object']

catOH_cols = [col for col in caterical_cols if x_train[col].nunique() > 10]
catOr_cols = [col for col in caterical_cols if x_train[col].nunique() <= 10]

pre_processor = ColumnTransformer(transformers = [
    ('num_process',SimpleImputer(strategy = 'constant'),numerical_cols),
    ('catOH',Pipeline(steps = [
        ('missing',SimpleImputer(strategy = 'most_frequent')),
        ('encode',OneHotEncoder(handle_unknown = 'ignore'))
    ]),catOH_cols),
    ('catOr',Pipeline(steps = [
        ('missing',SimpleImputer(strategy = 'most_frequent')),
        ('encode',OrdinalEncoder(handle_unknown = 'error')) 
    ]),catOr_cols)
])
```
